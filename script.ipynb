```python
from datetime import datetime
import pandas as pd


df =pd.read_excel('prov.xlsx', sheet_name='AJ2')

# Funktion zum Erstellen der PICA-Daten
def pica_schreiben(row):
    pica = ""  # Initialisiere die PICA-Variable
    
    # Prüfen, ob eine der Spalten in Gruppe 1 einen Wert enthält
    if any(pd.notna(value) for value in [row['1idn'], row['1prov'], row['1stempel'], row['1exlibris'], row['1auto'], row['1signatur'], row['1num']]):
        pica += f"6800 [Provenienz]\n"

    if pd.notna(row['1idn']):
        pica += f"6800 {row['1idn']}\n"

    if pd.notna(row['1prov']) and pd.isna(row['1idn']):
        pica += f"6800 {row['1prov']}\n"

    if pd.notna(row['1stempel']):
        pica += f"6800 Stempel *{row['1stempel']}\n"

    if pd.notna(row['1exlibris']):
        pica += f"6800 Exlibris *{row['1exlibris']}\n"

    if pd.notna(row['1auto']):
        pica += f"6800 Autogramm *{row['1auto']}\n"

    if pd.notna(row['1signatur']):
        pica += f"6800 Altsignatur *{row['1signatur']}\n"

    if pd.notna(row['1num']):
        pica += f"6800 Zugangsnummer *{row['1num']}\n"

    # Prüfen, ob eine der Spalten in Gruppe 2 einen Wert enthält
    if any(pd.notna(value) for value in [row['2idn'], row['2prov'], row['2stempel'], row['2exlibris'], row['2auto'], row['2signatur'], row['2num']]):
        pica += f"6801 [Provenienz]\n"

    if pd.notna(row['2idn']):
        pica += f"6800 {row['2idn']}\n"

    if pd.notna(row['2prov']) and pd.isna(row['2idn']):
        pica += f"6800 {row['2prov']}\n"

    if pd.notna(row['2stempel']):
        pica += f"6800 Stempel *{row['2stempel']}\n"

    if pd.notna(row['2exlibris']):
        pica += f"6800 Exlibris *{row['2exlibris']}\n"

    if pd.notna(row['2auto']):
        pica += f"6800 Autogramm *{row['2auto']}\n"

    if pd.notna(row['2signatur']):
        pica += f"6800 Altsignatur *{row['2signatur']}\n"

    if pd.notna(row['2num']):
        pica += f"6800 Zugangsnummer *{row['2num']}\n"

    # Zusätzliche Prüfung für `nummern`
    if pd.notna(row['nummern']):
        # Überprüfe, ob `6800`, `6801` oder `6802` bereits vorhanden sind
        if '6800' not in pica:
            pica += f"6800 [Provenienz]\n6800 unbekannte Herkunft\n6800 Nummern* {row['nummern']}\n"
        elif '6801' not in pica:
            pica += f"6801 [Provenienz]\n6801 unbekannte Herkunft\n6801 Nummern* {row['nummern']}\n"
        else:
            pica += f"6802 [Provenienz]\n6802 unbekannte Herkunft\n6802 Nummern* {row['nummern']}\n"
    
    return pica

# Erstelle einen Dateipfad mit aktuellem Datum
now = datetime.now()
dateipfad = f"ergebnis_{now.strftime('%Y-%m-%d_%H-%M-%S')}.dat"

# Schreibe das Ergebnis in die .dat-Datei
with open(dateipfad, 'a', encoding='utf-8') as datei:  # Datei im Anhängemodus öffnen
    for _, row in df.sample(10).iterrows():  # Wähle 10 zufällige Zeilen aus
        pica_inhalt = pica_schreiben(row)  # Erstelle den PICA-Inhalt
        datei.write(pica_inhalt + "\n")  # Schreibe in die Datei

print(f"Die Ergebnisse wurden in der Datei '{dateipfad}' gespeichert.")
```
